```python
%matplotlib inline
import numpy as np
import pickle
from scipy.stats import norm, truncnorm
import matplotlib.pyplot as plt

print('Import done.')
```

```python
def sample():
    # Generate some data for this demonstration.
    data = norm.rvs(10.0, 2.5, size=500)
    print(data)

    # Fit a normal distribution to the data:
    mu, std = norm.fit(data)

    # Plot the histogram.
    plt.hist(data, bins=25, density=True, alpha=0.6, color='g')

    # Plot the PDF.
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mu, std)
    plt.plot(x, p, 'k', linewidth=2)
    title = "Fit results: mu = %.2f,  std = %.2f" % (mu, std)
    plt.title(title)

    plt.show()

sample()
```

Fit samples from the dataset using truncated norm distribution.

```python
def fitsome():
    xmin, xmax = 0, 11
    idx = np.linspace(xmin, xmax, 100)
    idy = np.linspace(0, 1, 100)
    with open('AVA//AVA_corrected.txt', 'r') as handle:
        orig = [sum([int(l) for l in line.split()[1:]][1:11]) for line in handle]
    with open('train.pickle', 'rb') as handle:
        annotations = pickle.load(handle)
    

    def fitone(img_id): # untruncated
        annotation = annotations[img_id][2:12]
        score = annotations[img_id][12]
        
        # Recovered votes
        data = []
        for i in range(10):
            cnt = int(annotation[i]*orig[img_id])
            data += [i+1 for _ in range(cnt)]
        print(data)

        # mean, standard deviation
        mean = score
        std = (((np.array(data) - mean) ** 2).mean()) ** 0.5
        print(mean, std, np.std(data))
        
        # fit norm and trunc
        fit_mu, fit_std = norm.fit(data)
        print("Diff form fit: mean %.4f std %.4f %.4f %.4f" % (mean-fit_mu, std-fit_std, np.mean(data)-fit_mu, np.std(data)-fit_std))

        # gen pdf
        a, b = (1 - mean) / std, (10 - mean) / std
        p_trunc = truncnorm.pdf(idx, a, b, loc = mean, scale = std)
        p = norm.pdf(idx, mean, std)
        
        # plot score
        plt.plot([score for _ in range(100)], idy, 'r--')
        # plot vote as scattered points
        plt.plot(np.arange(1, 11, 1), annotation, 'x')
        # plot truncated
        plt.plot(idx, p_trunc, 'blue')
        # plot norm
        plt.plot(idx, p, 'k--', linewidth=1)

        title = "(%d) score = %.2f,  std = %.2f" % (img_id, mean, std)
        plt.title(title)
        plt.show()
    
    fitone(0)
    # exit()
    fitone(48633)
    # plt.show()
    # fitone(101783)
    # plt.show()
    # fitone(156972)
    # plt.show()

    # fitone(1156)
    # plt.show()
    # fitone(6534)
    # plt.show()
    # fitone(7099)
    # plt.show()

    # return None
fitsome()
        
```

```python
import numpy as np
import pickle
from scipy.stats import norm, skewnorm, truncnorm
from tqdm import tqdm

with open('AVA//AVA_corrected.txt', 'r') as handle:
    vote_count = [sum([int(l) for l in line.split()[1:]][1:11]) for line in handle]
```

```python
with open('train.pickle', 'rb') as handle:
        annotations = pickle.load(handle)
new_annotations = np.zeros((len(annotations), 4))
for idx in tqdm(range(len(annotations))):
    new_annotations[idx][0:2] = annotations[idx][0:2]
    
    img_id = int(annotations[idx][0])
    annotation = annotations[idx][2:12]
    data = []
    for i in range(10):
        cnt = int(annotation[i]*vote_count[img_id])
        data += [i+1 for _ in range(cnt)]
    
    mu, std = norm.fit(data)
    new_annotations[idx][2] = mu
    new_annotations[idx][3] = std
with open('new_train.pickle', 'wb+') as f:
    pickle.dump(np.array(new_annotations), f)
    
```

```python
with open('test.pickle', 'rb') as handle:
        annotations = pickle.load(handle)
new_annotations = np.zeros((len(annotations), 4))
for idx in tqdm(range(len(annotations))):
    new_annotations[idx][0:2] = annotations[idx][0:2]

    img_id = int(annotations[idx][0])
    annotation = annotations[idx][2:12]
    data = []
    for i in range(10):
        cnt = int(annotation[i]*vote_count[img_id])
        data += [i+1 for _ in range(cnt)]

    mu, std = norm.fit(data)
    new_annotations[idx][2] = mu
    new_annotations[idx][3] = std
with open('new_test.pickle', 'wb+') as f:
    pickle.dump(np.array(new_annotations), f)
```
